# vis for age

```python
age_wise = cancer1[['AGE','YY']]
```

```python
def age(x):
    if 0 < x <= 14:
        return '0-14'
    elif 14 < x <= 34:
        return '15-34'
    elif 34 < x <= 64:
        return '35-64'
    else:
        return '65+'
```

```python
age_wise['category'] = age_wise['AGE'].apply(age)
```

```python
age_wise.drop('AGE',1,inplace = True)
```

```python
age_wise1 = age_wise.sort_values('category')[0:8].groupby('YY').count()
age_wise2 = age_wise.sort_values('category')[8:463].groupby('YY').count()
age_wise3 = age_wise.sort_values('category')[463:4136].groupby('YY').count()
age_wise4 = age_wise.sort_values('category')[4136:].groupby('YY').count()
```

```python
fig = plt.figure(figsize = (13,6))
plt.plot(age_wise1,'b')
plt.plot(age_wise2,'r')
plt.plot(age_wise3,'g')
plt.plot(age_wise4,'k')
plt.title('Age wise Cancer count',fontsize = 30,color= 'w',backgroundcolor = 'k')
plt.xlabel('Year',fontsize = 20)
plt.ylabel('Cancer Count',fontsize = 20)
plt.legend(['0-14','15-34','35-64','65+'],fontsize = 16)
fig.show()
```

# Time series

```python
mnt = cancer1["DIAGDF"].astype(str)
```

```python
mnt
```

```python
lst2 = list()
for i in mnt:
    lst2.append(i[:7])
```

```python
month = dict()
for i in lst2:
    month[i] = month.get(i,0) + 1
    
```

```python
mth = pd.DataFrame()
```

```python
mth['Mnth'] = month.keys()
mth['Cancer_count'] = month.values()
```

```python
mth
```

```python
def dat(x):
    y = str(x)
    return y + '-01'
```

```python
mth['Mnth'] = mth['Mnth'].apply(dat)
```

```python
mth
```

```python
mth['Mnth'] = pd.to_datetime(mth['Mnth'])
```

```python
mth.set_index('Mnth',inplace = True)
```

```python
mth
```

```python
mth.to_csv('trend_timeseries.csv')
```

```python
mth.plot()
```

```python
mth.index
```

# time series--peri

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(mth['Cancer_count'],order=(1,1,1))
model_fit=model.fit()
```

```python
model_fit.summary()
```

```python
mth['forecast']=model_fit.predict(start=1,end=360,typ="levels")
mth[['Cancer_count','forecast']].plot(figsize=(12,8))
```

```python
from sklearn.metrics import mean_squared_error 
from statsmodels.tools.eval_measures import rmse
mean_squared_error(mth['Cancer_count'],mth["forecast"])
```
